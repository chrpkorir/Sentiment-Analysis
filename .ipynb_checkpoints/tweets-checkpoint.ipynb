# Data Cleaning

```python
# we first combine training and testing data so as to clean

combi = train.append(test,ignore_index=True)
combi.shape
```

# Steps in cleaning raw tweets data

1. Remove twitter handles as they are already marked @user for privacy concerns
2. Get rid of punctuation, numbers and special characters; they wouldn't help in     differentiating different types of tweets
3. Remove smaller words(stop words); they don't have much value
4. Normalize text data. Lemmatization

1. Remove twitter handle

```python
def remove_pattern(input_txt, pattern):
    r = re.findall(pattern, input_txt)
    for i in r:
        input_txt = re.sub(i, '', input_txt)
    return input_txt

combi['tidy tweet']= np.vectorize(remove_pattern)(combi['tweet'], "@[\w]*")
combi.head()
```

2. Remove Punctuation, Numbers and Special characters

```python
combi['tidy tweet']= combi['tidy tweet'].str.replace("[^a-zA-Z#]", "")
combi.head(10)
```

3. Remove stop words

```python
combi['tidy tweet'] = combi['tidy tweet'].apply(lambda x: ''.join([w for w in x.split() if len(w)>3]))
combi.head(10)
```

4. Text Normalization


```python
# Tokenize tweets

tokenized_tweet =  combi['tidy tweet'].apply(lambda x: x.split()) # tokenizing

tokenized_tweet.head()

```

```python
# Normalize the tokenized tweet

from nltk.stem.porter  import *
stemmer = PorterStemmer()

tokenized_tweet = tokenized_tweet.apply(lambda x: [stemmer.stem(i) for i in x]) # stemming

# stitch the tokens together

for i in range(len(tokenized_tweet)):
    tokenized_tweet[i] = ''.join(tokenized_tweet[i])

combi['tidy tweet'] = tokenized_tweet    
```

```python
# understand the impact of twitter handles in our dataset

def hashtag_extract(x): # function to collect hashtags
    hashtags = []
    #loop ove the words in tweet
    for i in x:
        ht = re.findall(r"#(w/+)", i )
        hashtags.append(ht)
        return hashtags
        
# extract hashtags from non racist/sexist tweet        
HT_regular = hashtag_extract(combi['tidy tweet'][combi['label'] ==0])


# extract hasgtags from racust/sexist tweets
HT_negative = hashtag_extract(combi['tidy tweet'][combi['label'] ==1])

# unnest list
HT_regular = sum(HT_regular,[])

HT_negative = sum(HT_negative,[])
```

```python
# check the hashtags in non racust/sexist tweets

a = nltk. FreqDist(HT_regular)
d = pd.DataFrame({'Hashtag' : list(a.keys()), 'Count': list(a.values())})

d = d.nlargest(columns= "Count", n=20)
plt.figure(figsize= (13, 6))
ax = sns.barplot(data=d, x ="Hashtag", y = "Count")
ax.sns(ylabel='Count')
plt.show()
```
